2) Selección de registros tomados en cali: 

```python
# ==========================================
# 1. CARGAR MAPAS Y CONSTRUIR GEOMETRÍAS

print("Cargando el shapefile...")
ruta_shapefile = r'C:\MaestriaCienciasDatos\Semestre3\Proyecto2\data\cali_completo_2025\cali.shp'
mapa_cali = gpd.read_file(ruta_shapefile).to_crs("EPSG:4326")

# Extraer la Comuna y fusionar los polígonos
mapa_cali['comuna'] = mapa_cali['id_sector'].astype(str).str.zfill(4).str[:2]
print("Construyendo polígonos de Comunas...")
mapa_comunas = mapa_cali.dissolve(by='comuna').reset_index()

# Polígono combinado (Límite municipal para validación)
muni = mapa_comunas.unary_union  

# ==========================================
# 2. PREPARAR DATASET DE DENGUE
print("Validando coordenadas y creando geometrías...")
# Filtrar coordenadas válidas
coords_ok = np.isfinite(df_dengue["x"]) & np.isfinite(df_dengue["y"])

# Crear GeoSeries solo con coordenadas válidas
geo = gpd.GeoSeries(
    gpd.points_from_xy(df_dengue.loc[coords_ok, "x"], df_dengue.loc[coords_ok, "y"]),
    index=df_dengue.index[coords_ok],
    crs="EPSG:4326"
)

# ==========================================
# 3. VALIDACIÓN GEOESPACIAL (Dentro vs Fuera)
# Flag dentro de Cali (incluye borde). Rellenar False donde no hay coordenada.
df_dengue["en_cali"] = geo.intersects(muni).reindex(df_dengue.index, fill_value=False).astype(bool)

df_in  = df_dengue[df_dengue["en_cali"]].copy()
df_out = df_dengue[~df_dengue["en_cali"]].copy()

print("\n--- RESUMEN DE VALIDACIÓN ESPACIAL ---")
print(df_dengue["en_cali"].value_counts())
print(f"Dentro: {len(df_in)} ({len(df_in)/len(df_dengue):.2%}) | "
      f"Fuera: {len(df_out)} ({len(df_out)/len(df_dengue):.2%})\n")

# ==========================================
# 4. GRÁFICO DE VALIDACIÓN
print("Generando mapa de validación visual...")
# Crear GeoDataFrame base para el ploteo
gdf = gpd.GeoDataFrame(df_dengue.loc[coords_ok], geometry=geo, crs="EPSG:4326")
 
fig, ax = plt.subplots(figsize=(10, 10))

# Dibujar Límite Municipal y las divisiones de las Comunas
mapa_comunas.boundary.plot(ax=ax, linewidth=1, color="black", alpha=0.9, zorder=1) # Bordes comunas
gpd.GeoSeries([muni]).boundary.plot(ax=ax, linewidth=1.5, color="black", alpha=0.9, zorder=2) # Borde municipio

# Puntos Dentro y Fuera
gdf[gdf["en_cali"]].plot(ax=ax, markersize=8, alpha=0.8, zorder=3, label="Dentro", color="C0")
gdf[~gdf["en_cali"]].plot(ax=ax, markersize=8, alpha=0.8, zorder=4, label="Fuera", color="red")

# Leyenda personalizada
legend_elements = [
    Line2D([0], [0], marker='s', color='w', label='Dentro de Cali', markerfacecolor='C0', markersize=8),
    Line2D([0], [0], marker='s', color='w', label='Fuera de Cali', markerfacecolor='red', markersize=8),
    Line2D([0], [0], color='black', lw=1.5, label='Limitación Comunas')
]
ax.legend(handles=legend_elements, loc='upper left', frameon=True, edgecolor='black', facecolor='white', fontsize=10)

ax.set_title("Validación geoespacial de puntos respecto al límite municipal de Cali", fontsize=14, fontweight='bold')
ax.set_xlabel("Longitud")
ax.set_ylabel("Latitud")
ax.set_aspect('equal', adjustable='datalim')
ax.grid(True, linestyle="--", linewidth=0.3, alpha=0.5)
plt.tight_layout()
plt.show()

# ==========================================
# 5. CRUCE ESPACIAL Y GUARDADO (Spatial Join)

print("\nEjecutando cruce espacial definitivo para extraer comunas...")
# Hacemos el sjoin usando solo las comunas y la geometría
dataset_final = gpd.sjoin(gdf, mapa_comunas[['nombre','comuna', 'geometry']], how="left", predicate="within")

if 'index_right' in dataset_final.columns:
    dataset_final = dataset_final.drop(columns=['index_right'])

# --- Filtrar datos de cali ---
print("Filtrando para conservar ÚNICAMENTE los registros dentro de Cali...")
dataset_final = dataset_final[dataset_final["en_cali"] == True].copy()

# Validar Nulos en Comuna
nulos_comuna = dataset_final['comuna'].isnull().sum()
print(f"Cruce finalizado. Registros totales a modelar (Dentro de Cali): {len(dataset_final)}")
# Si quedan nulos aquí, son casos muy raros justo en la frontera del polígono
print(f"Registros dentro de Cali pero sin comuna asignada: {nulos_comuna}") 

# --- Creación automatica de carpeta y guardado ---
print("\nPreparando el guardado del dataset actualizado...")
df_final_export = pd.DataFrame(dataset_final.drop(columns=['geometry', "en_cali"]))

# Definimos la ruta usando pathlib
ruta_guardado = r"C:\MaestriaCienciasDatos\Semestre3\Proyecto2\data\procesado\dataset_dengue_con_comuna.parquet"
ruta_path = Path(ruta_guardado)

# Creamos la carpeta padre ("procesado") si no existe
ruta_path.parent.mkdir(parents=True, exist_ok=True)

# Guardamos el archivo final
df_final_export.to_parquet(ruta_path, index=False)
print(f"¡Proceso exitoso! Archivo creado y guardado en:\n{ruta_path}")
```

```python

# Extraemos los 20 colores de tab20
colores_tab20 = plt.get_cmap('tab20').colors
# Extraemos colores complementarios de Set3
colores_set3 = plt.get_cmap('Set3').colors

# Unimos las dos listas de colores (tendremos más de 30 colores distintos)
colores_combinados = colores_tab20 + colores_set3

# Creamos nuestro mapa de colores personalizado
paleta_q1 = ListedColormap(colores_combinados)
# ==========================================

fig, ax = plt.subplots(figsize=(8, 8))

# 1. Capa base: Graficar los barrios (sectores), coloreados por comuna
mapa_cali.plot(
    ax=ax,
    column='comuna',
    cmap=paleta_q1,       # <-- Emplear paleta creada
    edgecolor='white',    # Líneas blancas muy finas para ver los límites de los barrios
    linewidth=0.3, 
    alpha=0.8,            # Claridad de los colores
    legend=False 
)

# 2. Capa intermedia: Resaltar las fronteras de las comunas
mapa_comunas.boundary.plot(
    ax=ax,
    color='black',
    linewidth=1.2,
    zorder=2              # Asegura que las fronteras se dibujen sobre los barrios
)

# 3. Capa superior: Etiquetas numéricas de las comunas
for idx, row in mapa_comunas.iterrows():
    # representative_point() es más seguro que centroid()
    punto = row.geometry.representative_point()
    
    ax.annotate(
        text=str(row['comuna']),
        xy=(punto.x, punto.y),
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=11,
        fontweight='bold',
        color='black',
        # Agregamos caja de fondo para que el número resalte sin importar el color de fondo
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2', alpha=0.85),
        zorder=3
    )

# ==========================================
# 4. NUEVO: LEYENDA MANUAL DE LÍMITES
# ==========================================
# Creamos la representación del límite de la comuna (Línea negra)
leyenda_comuna = Line2D([0], [0], color='black', linewidth=1.5, label='Límites de comunas')

# Creamos la representación del límite del barrio (Recuadro gris con borde blanco)
leyenda_barrio = Line2D([0], [0],color='white', linewidth=1.5, label='Límites de barrios')

# Insertamos la leyenda en el gráfico
ax.legend(
    handles=[leyenda_comuna, leyenda_barrio],
    loc='upper left',
    title_fontsize=12,
    fontsize=11,
    frameon=True,
    edgecolor='black',
    facecolor='lightgray',
    framealpha=0.6,
    borderpad=0.8
)
# ==========================================

# 5. Formato y estética del gráfico
ax.set_title("Definición de la unidad espacial de análisis: delimitación de comunas en Cali", fontsize=14, pad=15)
ax.set_xlabel("")
ax.set_ylabel("")

# --- Nota explicativa para la zona 81 ---
ax.text(
    0.02, 0.02, 
    "* Nota: El polígono 81 corresponde a la Zona de Expansión Urbana.",
    transform=ax.transAxes,
    fontsize=11,
    verticalalignment='bottom',
    horizontalalignment='left',
    bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='gray', alpha=0.9),
    zorder=4
)

# Mantiene la proporción geográfica real
ax.set_aspect('equal', adjustable='datalim')
ax.set_axis_off()  # Oculta los ejes para un mapa más limpio

ax.grid(False)
plt.savefig("mapa_comunas_cali.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
