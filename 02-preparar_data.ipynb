### Esquema de agregación por semana epidemiológica a nivel comuna (Cali)

| **Grupo de variables** | **Agregación (Píxel a Comuna)** | **Justificación técnica y espacial** | **Ejemplos de Columnas** |
|---|---|---|---|
| Epidemiología (Absoluta) y Conteos Físicos | `sum` | Los eventos de salud y los objetos físicos (criaderos/árboles) son discretos y aditivos en el espacio. Se suman todos los píxeles para obtener la carga o capacidad vectorial total de la comuna. | `casos`, `conteo`, `conteo_sumideros`, `n_arboles`, `n_pozos`, y todos sus rezagos (`conteo_lag*`) |
| Epidemiología (Relativa) | `mean` | Ante la ausencia de denominadores poblacionales temporales, la Tasa Empírica Bayesiana se promedia espacialmente. Esta heurística es válida ya que el estimador bayesiano conserva su corrección de varianza y suavizado original. | `tasa_eb_100000`, `tasa_eb_100000_lag*` y sus ventanas móviles |
| Clima: Precipitación (Volumétrica) | `sum` | Al sumar los milímetros de lluvia de todos los píxeles, se calcula el volumen total acumulado de agua en la jurisdicción, capturando la disponibilidad hídrica real para la formación de criaderos. | `precip_fore`, `precip_fore_lag*`, `precip_fore_sum_*`|
| Clima base, derivado | `mean` | Variables continuas de naturaleza intensiva. El promedio espacial captura de manera fiel la exposición termodinámica y climática representativa de toda la zona. | `tmax_fore`, `ndvi`, `ndwi`, `amplitud_termica`,`estres_hidrico` (y sus rezagos) |
| Eventos climáticos temporales (Rachas) | `mean` | Frecuencias temporales (ej. días > 26°C). Sumarlas en el espacio excede el límite temporal de la semana. Su promedio indica la intensidad media del evento en la comuna. | `tmax_mayor_26`, `racha_precip_then_tmax_6w` (y sus rezagos) |
| Estadísticas derivadas (Dispersion/Rangos) | `mean` | Las métricas estadísticas de distribución (desviación estándar, medias, mínimos, máximos) calculadas por píxel no son aditivas en el espacio y deben promediarse para mantener coherencia. | `*_sd_*`, `*_mean_*`, `*_max_*`, `*_min_*`,(incluidas lluvias o conteos) |
| Topografía e Hidrología | `mean` | Atributos estructurales (áreas, distancias, relieves). El promedio espacial resume la morfología predominante y la exposición hídrica del territorio. | `slope`, `aspect`, `valle`, `area_humedal_m2`, `distancia_rio_m`, `distancia_pozos_m` |
| Uso del suelo y Sociodemográficas | `mean` | Variables de proporción y densidad. El promedio espacial mantiene la configuración urbana, estrato social y perfil poblacional característico de la comuna. | `comercial`, `habitacional`, `estrato`, `hacinamien`, `TP34_1_EDA` a `TP34_9_EDA` |
| Identificadores Espaciales | `Excluir` | Variables georreferenciadas a nivel de píxel pierden sentido al colapsar el área hacia un polígono administrativo. | `x`, `y`, `pixel_id`, `spatial_ref`, `band` |

#### Consideraciones estadísticas y salvaguardas en la agregación espacial

El proceso de agrupación de una malla de píxeles a límites de comuna requiere reglas matemáticas estrictas que van más allá de la división por dominios temáticos. Para garantizar la integridad del conjunto de datos multidimensional, se implementaron las siguientes salvaguardas metodológicas:

1) Manejo Dual de la Precipitación y sus Derivados: Aunque la precipitación pura y sus acumulados semanales (precip_fore, precip_fore_sum_4w) se agregan mediante suma (sum) para reflejar el volumen de agua captado por la comuna, las estadísticas derivadas de la misma lluvia (como la desviación estándar _sd_, los promedios móviles _mean_ o los valores máximos _max_) se agrupan estrictamente mediante promedio espacial (mean). Sumar varianzas o promedios en el espacio destruye la distribución estadística de la variable, por lo que el promedio espacial conserva la métrica de dispersión representativa del territorio.

2) Tratamiento de Eventos Temporales (Rachas): Las variables que contabilizan la frecuencia temporal de un evento (ej. tmax_mayor_26 o racha_precip_then_tmax_6w) no son aditivas en el espacio. Si una semana tiene 7 días, sumar las rachas de cientos de píxeles arrojaría resultados matemáticamente imposibles (ej. "700 días de calor" en una semana). Por tanto, se aplica un promedio espacial (mean), interpretándose como "la comuna experimentó un promedio de $X$ días con la condición específica".

3) Preservación del Suavizado Bayesiano: Como lineamiento operativo, las variables de incidencia relativa (Tasa de Incidencia Empírica Bayesiana - TIEB) se agregan mediante el promedio. Aunque tradicionalmente las tasas brutas requieren recálculo con denominadores poblacionales, el uso de la TIEB permite esta heurística, dado que el estimador ya penaliza intrínsecamente la inestabilidad de las varianzas en áreas pequeñas, transfiriendo este suavizado al valor promedio de la comuna.

```python
print("Iniciando agregación espacial: Píxel -> Comuna...")

# 1. Definir llaves y variables a descartar
group_cols = ["comuna", "epi_year", "epi_week", "epi_week_number"]

# Excluimos variables de píxeles que ya no nos sirven
exclude_cols = ["x", "y", "pixel_id", "spatial_ref", "band", "nombre"]

# 2. Motor de clasificación de columnas 
sum_cols = []
mean_cols = []

# Palabras clave de ESTADÍSTICAS derivadas que NUNCA deben sumarse en el espacio
estadisticas_no_sumables = ["mean", "sd", "max", "min", "racha"]

for col in df.columns:
    # A. Saltar las llaves y las columnas excluidas
    if col in group_cols or col in exclude_cols:
        continue
        
    # B. Grupo Absoluto Directo (SUMA segura)
    if col in ["casos", "n_arboles", "n_pozos"]:
        sum_cols.append(col)
        
    # C. Familias de Conteos (SUMA) - Excluyendo sus medias y desviaciones
    elif "conteo" in col and not any(sub in col for sub in estadisticas_no_sumables):
        sum_cols.append(col)
        
    # D. Familias de Precipitación (SUMA volumétrica) - Excluyendo sus medias, desviaciones y rachas
    # Nota: Los 'precip_fore_sum_*' SI pasarán este filtro y se sumarán.
    elif "precip" in col and not any(sub in col for sub in estadisticas_no_sumables):
        sum_cols.append(col)
        
    # E. Todo lo demás: Tasa EB, Clima base, Topografía, Uso de suelo, y las Estadísticas (PROMEDIO)
    else:
        mean_cols.append(col)


# 3. Construir diccionario y tratar Nulos
# Reemplazamos el "sum" simple por una función que exige al menos 1 dato real para no sesgar datos falsos ya que sum por defecto si son nan los trata como ceros, lo que no es correcto en este caso.
agg_dict = {col: lambda x: x.sum(min_count=1) for col in sum_cols}
agg_dict.update({col: "mean" for col in mean_cols})

print(f"Plan de agregación validado:")
print(f" - {len(sum_cols)} variables a SUMA (con protección contra falsos ceros).")
print(f" - {len(mean_cols)} variables a PROMEDIO espacial/estadístico.")

# ==========================================
# 4. Ejecutar Agregación (Downgrading espacial)
print("\nCalculando agrupaciones comunales...")
df_comuna_semana = (
    df
    .groupby(group_cols, as_index=False)
    .agg(agg_dict)
)

# 5. Ordenar cronológicamente
df_comuna_semana = df_comuna_semana.sort_values(by=["comuna", "epi_year", "epi_week_number"]).reset_index(drop=True)

# 6. Guardar el resultado final
ruta_guardado = r"C:\MaestriaCienciasDatos\Semestre3\Proyecto2\data\procesado\dataset_comuna_semana.parquet"
df_comuna_semana.to_parquet(ruta_guardado, index=False)

print(f"\n¡Transformación exitosa!")
print(f"Dimensiones del nuevo dataset: {df_comuna_semana.shape} (Filas: Comuna-Semana)")
print(f"Archivo guardado y listo para Fase 4 en: \n{ruta_guardado}")
```
